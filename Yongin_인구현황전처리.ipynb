```python
import pandas as pd
```

```python
df=pd.read_excel('★(용인시)연령별 인구 현황(11월말 기준).xlsx',
                 sheet_name=0,
                 header=2)
```

```python
df.head()
```

```python
col_list = []
for col in df.columns[1:]:
    if 'Unnamed' not in col:
        col_list.append(col)
        
col_list=[col.replace(' ','') for col in col_list]
```

```python
check_list =[]
for col in col_list:
    for word in [' 계',' 남',' 여']:
        w = col + word
        check_list.append(w)
```

```python
check_list.insert(0,'연령별')
```

```python
df.columns = check_list
```

```python
df.drop(0,inplace=True)
df.head()
```

집계용 데이터를 만들어 주자.

```python
order_list=df[(df['연령별'].str.contains('대')==True)|(df['연령별'].str.contains('110')==True)]['연령별'].values
```

```python
input_list=[]
for state in col_list:
    for age in order_list:
        for mvw in ['남','여']:
            value_list= []
            value_list.append(state)
            value_list.append(age)
            value_list.append(mvw)
            input_list.append(value_list)
```

```python
agg_df=pd.DataFrame(input_list,columns=['읍면동','연령','성별'])
agg_df.head()
```

```python
for idx in agg_df.index:
    agg_df.loc[idx,'인구수'] = df.loc[df['연령별'] == agg_df.loc[idx,'연령'],
                                   agg_df.loc[idx,'읍면동'] +' '+ agg_df.loc[idx,'성별']].values[0]
    
```

```python
agg_df.head()
```

```python
agg_df.to_csv('용인시 인구정보.csv',index=False,encoding='utf-8-sig')
```
